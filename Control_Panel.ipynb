# Dota 2 Pipeline Control Panel
Use this local notebook to trigger your containerized pipeline. Instead of running Python or PySpark heavily on your machine, these cells simply act as a remote control for your Docker cluster.

### 1. Set Match Count

```python
# Set how many matches to fetch per run (e.g., 20, 50, 100)
number_of_matches = 10
print(f"Pipeline configured to fetch {number_of_matches} matches per run.")
```

### 2. Run Data Collector
This pulls the most recent public matches from OpenDota, transforms them into player event logs, streams them to Kafka, and pushes raw logs to Elasticsearch.

```python
# Old command (Windows native):
# !docker-compose run --rm -e MATCH_BATCH_SIZE={number_of_matches} data-collector python main.py

# New command (Forces WSL execution):
# !wsl -e bash -c "cd /mnt/d/work/bigdata/dota2-analytics && docker compose run --rm -e MATCH_BATCH_SIZE={number_of_matches} data-collector python main.py"

import subprocess

cmd = f"docker-compose run --rm -e MATCH_BATCH_SIZE={number_of_matches} data-collector python main.py"
print(f"Executing: {cmd}")
subprocess.run(cmd, shell=True)

```

### 3. Run Spark Processor
This consumes the Kafka stream, aggregates win rates and statistics by hero using PySpark, and outputs the final datasets to Hadoop HDFS and Elasticsearch.

```python
# Old command (Windows native):
# !docker-compose run --rm spark-processor

import subprocess
cmd = "docker-compose run --rm spark-processor"
print(f"Executing: {cmd}")
subprocess.run(cmd, shell=True)

# New command (Forces WSL execution):
# !wsl -e bash -c "cd /mnt/d/work/bigdata/dota2-analytics && docker compose run --rm spark-processor"
```
